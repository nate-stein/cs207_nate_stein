```python
import numpy as np
x = np.linspace(0,25)
y1 = [2*i for i in x]
y2 = [pow(i, 2) for i in x]
y3 = [i+5 for i in x]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.title('Various Functions', fontsize=24)
plt.plot(x, y1, lw=4, ls='-', label=r'$2^x$')  # linewidth = 4, linestyle = solid, raw string label
plt.plot(x, y2, lw=4, ls='-', label=r'$x^2$')
plt.plot(x, y3, lw=4, ls='-', label=r'$x+5$')
```
